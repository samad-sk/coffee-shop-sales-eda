```python
# Convert transaction_date and transaction_time to datetime datatype. 
sales_clean['transaction_date'] = pd.to_datetime(sales_clean['transaction_date'],errors='coerce')
sales_clean['transaction_time'] = pd.to_datetime(sales_clean['transaction_time'], format='%H:%M:%S').dt.time
```

# 9. Data Aggregation for Business Problems

#### **Questions**
1. Which product category has the highest number of units sold ? 
2. What is the total average minimum and maximum revenue generated by each product category ? 
3. Which store location generates the highest revenue and handles the most orders ?
4. Which product category drives the most revenue in each store location ?
5. What is the overall statistical summary of all numerical columns across each store location?

```python
# 1. Which product category has the highest number of units sold ? 
sales_clean.groupby('product_category')['quantity'].sum().reset_index().sort_values(by='quantity',ascending=False)

```

```python
# 2. What is the total average minimum and maximum revenue generated by each product category ? 
sales_clean['revenue']= sales_clean['quantity']*sales_clean['unit_price']
```

```python
sales_clean.groupby('product_category')['revenue'].agg(['sum','min','max','mean']).round(1).sort_values(by='sum',ascending=False)

# Insight : Coffee beans have the highest average revenue per transaction ($22.9) despite being 6th in total revenue — meaning low volume but high value purchases.
```

```python
# 3. Which store location generates the highest revenue and handles the most orders ? 
top_store = sales_clean.groupby('store_location').agg({
    'quantity':'sum',
    'revenue':'sum',
    'transaction_id': 'count'

})
top_store.sort_values(by='revenue',ascending=False).reset_index()

# Insight : All 3 stores perform almost equally — similar sales, revenue and transactions. No single store is significantly ahead or behind.
```

```python
sales_clean.columns
```

```python
# 4. Which product category drives the most revenue in each store location ? 
sales_clean.groupby(['store_location','product_category'])['revenue'].sum().reset_index().sort_values(by=['store_location','revenue'],ascending=[True,False])
# Insight : Coffee drives the revenue in every store ,
#          Packaged chocolate and Flavors are the lowest revenue categories everywhere. 
```

# 2. Analysis using Matplotlib 
**1. Which product category do customers buy the most?**
- Insight : Coffee and Tea together account for over 69% of all units sold
- Recommendation : Introduce seasonal coffee and tea variants to maintain demand. Bundle them with low selling items like Packaged Chocolate and Flavours to boost their sales.

**2. Which store location generates the most revenue?**
- Insight : All 3 stores perform almost equally — similar sales, revenue and transactions. No single store is significantly ahead or behind.
- Recommendation : Focus on reducing operation costs to improve profit margin. Consistent performance in store model is ideal for expanding to new locations.

**3. How did revenue change month over month?**
- Insight : Revenue grew steadily from January to June, peaking around May-June. The  drop in December is due to incomplete data — only 6 days of December are recorded.
- Recommendation : Run special offers before May-June to maximize peak sales. During slow months, introduce discounts and repeat customer deals to maintain revenue.

```python
# 1. Which product category do customers buy the most ? 
data = sales_clean.groupby('product_category')['quantity'].count().reset_index().sort_values(by='quantity',ascending=True)
```

```python
plt.barh(data['product_category'],data['quantity'] , color='orange')
plt.title("Which product category do customers buy the most ? ")
plt.xlabel('Total Quantity')
for index, value in enumerate(data['quantity']):
    plt.text(value, index, str(value))
plt.show()

# Insight :  Coffee and Tea together account for over 69% of all units sold 
```

```python
# 2. Which store location generates the most revenue?
revenue_by_store = sales_clean.groupby('store_location')['revenue'].sum().reset_index().sort_values(by='revenue',ascending=True)

plt.figure(figsize=(10, 5))
plt.bar(revenue_by_store['store_location'],revenue_by_store['revenue'],color='green')
plt.title("Which Store location generates the most revenue ?")
plt.ylabel("Revenue")
for index, value in enumerate(revenue_by_store['revenue']):
    plt.text(index, value, str(round(value,1)), ha='center')
plt.show()


# Insight : All 3 stores perform almost equally — similar sales, revenue and transactions. No single store is significantly ahead or behind.
```

```python
sales_clean['month'] = sales_clean['transaction_date'].dt.month_name()
```

```python
revenue_trend = sales_clean.groupby('month')['revenue'].sum().reset_index()

month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
revenue_trend['month'] = pd.Categorical(revenue_trend['month'], categories=month_order, ordered=True)
revenue_trend = revenue_trend.sort_values('month')
```

```python
plt.figure(figsize=(10, 5))  
plt.plot(revenue_trend['month'],revenue_trend['revenue'] , color='red' , marker='o')
plt.title("How did revenue change month over month?")
plt.tight_layout()
plt.plot()

# Insight - Revenue grew steadily from January to June, peaking around May-June. The  drop in December is due to incomplete data — only 6 days of December are recorded.
```

# 3. Analysis using Seaborn 
**1. Which store is most active ?**
- Insight : All stores are equally active and receive similar number of orders. No single store dominates in customer footfall. 
- Recommendation : Since data shows active customer demands. Business can expand and introduce branches in other locations.

**2. Which factors — quantity ordered or unit price — have a stronger impact on revenue?**
- Insight : Unit price has a stronger impact on revenue (0.69) than quantity ordered (0.36) — meaning selling higher priced products matters more than selling more items."
- Recommendation : Focus on promoting premium and higher priced products like coffee beans and branded items to increase revenue.

**3. What is the typical order size of a customer per transaction?**
- Insight : Most customers order only 1 or 2 items per transaction. 
- Recommendation : Introduce combo offers like 'Coffee + Snack' deals to encourage customers to buy more items per visit.

```python
# 1. Which store is most active ? 

sns.countplot(data=sales_clean, x='store_location' , color='purple')
plt.title("Which store is most active ? ")
plt.ylabel("Number of Transactions")
plt.show()
```

```python
# 2. Which factors — quantity ordered or unit price — have a stronger impact on revenue?
sns.heatmap(sales_clean[['quantity','unit_price','revenue']].corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Which factors — quantity ordered or unit price — have a stronger impact on revenue?")
plt.show()

# Insight : Unit price has a stronger impact on revenue (0.69) than quantity ordered (0.36) — meaning selling higher priced products matters more than selling more items."
```

```python

# 3. What is the typical order size of a customer per transaction?
plt.figure(figsize=(10,6))
sns.histplot(data=sales_clean, x='quantity', bins=10 , color='orange')
plt.title("How Many Items Do Customers Order Per Transaction?")
plt.show()

# Insight : Insight : Most customers order only 1 or 2 items per transaction. 
```
