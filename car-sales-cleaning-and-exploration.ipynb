# Fix column headers
- Convert to lower case and remove spaces
- Rename column names
- Fix data types 

```python
# Remove spaces , lower case from all column headers.  
Cars.columns = (
    Cars.columns
    .str.strip()
    .str.lower()
)
```

```python
# Check whether column headers got fixed
Cars.columns
```

```python
# Rename Column headers
Cars = Cars.rename(columns = {
    'sales_in_thousands':'sales',
    '__year_resale_value':'year_resale_value',
    'price_in_thousands':'price'
})
```

```python
# Multiply price and sales column by 1000 to fix the values.
Cars['price'] = Cars['price']*1000
Cars['sales'] = Cars['sales']*1000
```

### Data Types

- **manufacturer (object)**: Contains text values representing brand names. Object type is appropriate.
- **model (object)**: Contains text values for car models. Object type is suitable.
- **sales (float64)**: Numeric column representing number of units sold. Float is acceptable for calculations and analysis.
- **year_resale_value (float64)**: Numeric values representing resale value. Float type is appropriate for handling missing values and decimals.
- **vehicle_type (object)**: Categorical text data (e.g., passenger, SUV). Object type is suitable.
- **price (float64)**: Numeric column representing car price. Float is appropriate for calculations and comparisons.
- **engine_size (float64)**: Numeric measurement of engine capacity. Float type is suitable.
- **horsepower (float64)**: Numeric performance metric. Float type is appropriate.
- **wheelbase (float64)**: Numeric measurement of vehicle dimensions. Float is suitable.
- **width (float64)**: Numeric column representing car width. Float is appropriate.
- **length (float64)**: Numeric column representing car length. Float is appropriate.
- **curb_weight (float64)**: Numeric value representing vehicle weight. Float type is suitable.
- **fuel_capacity (float64)**: Numeric measurement of fuel tank capacity. Float is appropriate.
- **fuel_efficiency (float64)**: Numeric value representing efficiency. Float is suitable for analysis.
- **latest_launch (object → datetime)**: Currently stored as object but represents dates. Should be converted to datetime for proper time-based analysis.
- **power_perf_factor (float64)**: Numeric performance-related metric. Float type is appropriate.

```python
# Convert data type of latest_launch from object to datetime
Cars['latest_launch'] = pd.to_datetime(Cars['latest_launch'])
```

# Detecting Nulls
- We will count nulls for each column
- Data Imputation

```python
Cars.isnull().sum()
```

```python
# Show rows where year_resale_value is null 
((Cars.isnull().sum() / len(Cars)) * 100).round(1)
```

```python
# Fill the columns with mean()
cols = ['price', 'engine_size', 'horsepower', 'wheelbase', 'width',
        'length', 'curb_weight', 'fuel_capacity', 'fuel_efficiency',
        'power_perf_factor']

for col in cols:
    Cars[col] = Cars[col].fillna(Cars[col].mean())
```

```python
# Filling with medians in year_resale_value column
Cars['year_resale_value'] = Cars['year_resale_value'].fillna(Cars['year_resale_value'].median())
```

```python
# Checking the count of nulls 
Cars.isnull().sum()
```
